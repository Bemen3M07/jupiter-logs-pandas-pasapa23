```python
# Importem les llibreries necess√†ries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Carreguem les dades del fitxer CSV
df = pd.read_csv('estudiantes_notas.csv')
```

```python
# Calcular la mitjana de cada alumne (M01, M04, M03, M05)
df['Mitjana Alumne'] = df[['M01', 'M04', 'M03', 'M05']].mean(axis=1)
```

```python
# Generar gr√†fic 1: Gr√†fic de barres de la mitjana de cada alumne
plt.figure(figsize=(10,6))
plt.bar(df['Alumne'], df['Mitjana Alumne'], color='skyblue')
plt.title('Mitjana de cada alumne')
plt.xlabel('Alumne')
plt.ylabel('Mitjana')
plt.xticks(rotation=45)
plt.show()
```

```python
# Generar gr√†fic 2: Gr√†fic de l√≠nia de les notes mitjanes de tots els alumnes
plt.figure(figsize=(10,6))
plt.plot(df['Alumne'], df['Mitjana Alumne'], marker='o', color='orange', linestyle='-', linewidth=2)
plt.title('Evoluci√≥ de la mitjana per alumne')
plt.xlabel('Alumne')
plt.ylabel('Mitjana')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Generar gr√†fic 3: Gr√†fic de sectors per percentatge d'aprovats i suspesos
percentatge_aprovats = df['Mitjana Alumne'] >= 5
labels = ['Aprovats', 'Suspesos']
sizes = [percentatge_aprovats.sum(), len(df) - percentatge_aprovats.sum()]
colors = ['lightgreen', 'salmon']

plt.figure(figsize=(7,7))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Percentatge d\'Aprovats vs Suspensos')
plt.axis('equal')
plt.show()
```

```python
# Generar gr√†fic 4: Gr√†fic de barres per les notes m√©s altes i m√©s baixes de cada mat√®ria
m√†xims = df[['M01', 'M04', 'M03', 'M05']].max()
m√≠nims = df[['M01', 'M04', 'M03', 'M05']].min()

plt.figure(figsize=(10,6))
plt.bar(m√†xims.index, m√†xims, color='lightcoral', label='Nota m√†xima')
plt.bar(m√≠nims.index, m√≠nims, color='lightblue', label='Nota m√≠nima', alpha=0.7)
plt.title('Notes m√†ximes i m√≠nimes per mat√®ria')
plt.xlabel('Mat√®ria')
plt.ylabel('Nota')
plt.legend()
plt.show()
```
